```python
Scenario 2: A data analysis tool processes a list of numbers and needs to identify the most frequently occurring value.
 Write logic to find the most frequently occurring number in a given list.
```

```python
def mostFrequent(num):
    
    frequency_map = {}

    for i in num:
        if i in frequency_map:
            frequency_map[i] += 1
        else:
            frequency_map[i] = 1
        
    most_frequent_num = None
    max_freq = 0

    for number, freq in frequency_map.items():
        if freq > max_freq:
            max_freq = freq
            most_frequent_num = number
    print("MAP: ", frequency_map)
    return most_frequent_num

data1 = [1, 2, 3, 4, 1, 2, 2, 3, 1, 1, 4]
result1 = mostFrequent(data1)
print(f"Data: {data1}")
print(f"Most frequent number: {result1}")


```

```python

def checkAnagram(word1,word2):
    process_word1 = word1.replace(" " ,"").lower()
    process_word2 = word1.replace(" " ,"").lower()
    sort_word1 = sorted(process_word1)
    sort_word2 = sorted(process_word2)

    if sort_word1 == sort_word2:
        print(f"{word1} and {word2} are anagram")
    else:        
        print(f"{word1} and {word2} are  not anagram")

word1 = input("Enter word1: ")
word2 = input("Enter word2: ")
checkAnagram(word1,word2)
    
    

```

```python
4.Scenario: A speech analysis program needs to count the number of vowel sounds in a given input.
 Logic:
```

```python
def countVowel(msg):
    vowel = "AEIOUaeiou"
    count = 0
    for i in msg:
        if i in vowel:
            count +=1
    print("The total vowel count : ", count)

    
msg = input("Enter a message: ")
countVowel(msg)
```

```python
8.Scenario: A system needs to verify whether a given dataset contains duplicate entries.

```

```python
def duplicatecheck(dataset):
    value =set()
    for i in dataset:
        if i in value:
             return True        
        value.add(i)
    return False
dataset = [1,2,3,3,5,2,8]
duplicatecheck(dataset)
```

```python
9.Scenario: A digital calculator includes a feature to sum the digits of a number for verification purposes.
 Logic:
```

```python
10.Scenario: A language-learning app wants to verify whether a given sentence is a pangram (contains every letter of the alphabet at least once).
 
```

```python
def pangram(sentence):
    sen = sentence.lower()
    alpha = "".join([char for char in sen if char.isalpha()])
    print(alpha)
    letter = set()
    for i in alpha:
        if i not in letter:
            letter.add(i)
    print(letter)
    size =len(letter)
    if(size == 26):
        print("The given sentence is a pangram")
    else:
        print("The given sentence is not a pangram")
    
sentence ="The quick brown fox, jumps over the lazy dog!!"
pangram(sentence)
```
